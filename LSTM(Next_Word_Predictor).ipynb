```python
# necessary imports for nltk to work properly
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# tokenize
tokens = word_tokenize(document.lower())
```

# create vocab{}

```python
vocab = {"<UNK>" : 0}
```

```python
unique_tokens_with_count = Counter(tokens)
```

```python
unique_tokens = unique_tokens_with_count.keys()
unique_tokens = list(unique_tokens)
unique_tokens
```

```python
for token in unique_tokens:
  if token not in vocab:
    vocab[token] = len(vocab)
vocab
```

```python
len(vocab)
```

# Create a list of sentences

```python
list_of_sentences = document.split("\n")
len(list_of_sentences)
```

# Create a numerical list of sentences

```python
def sentence_to_indices(sentence, vocab):
  numerical_sentence = []
  for token in sentence:
    if token in vocab:
      numerical_sentence.append(vocab[token])
    else:
      numerical_sentence.append(vocab['<UNK>'])

  return numerical_sentence
```

```python
sentence_to_indices("You have to pay the entire fee of Rs 5600",vocab)
```

```python
list_of_numeric_sentences = []
for sentence in list_of_sentences:
  list_of_numeric_sentences.append(sentence_to_indices(word_tokenize(sentence.lower()),vocab))
len(list_of_numeric_sentences)
```

```python
list_of_numeric_sentences[:5]
```

# Create dataset

```python
dataset = []
for sentence in list_of_numeric_sentences:
  for i in range(1, len(sentence)):
    dataset.append(sentence[:i+1])
len(dataset)
```

```python
len_list = []
for sequence in dataset:
  len_list.append(len(sequence))
max(len_list)
```

```python
padded_sequences = []

for sequence in dataset:
  padded_sequences.append([0]*(max(len_list) - len(sequence)) + sequence)
len(padded_sequences[1])
```

```python
# Convert to tensor
padded_sequences = torch.tensor(padded_sequences, dtype = torch.long)
padded_sequences.shape
```

# Create X and Y

```python
X = padded_sequences[:, :-1]
y = padded_sequences[:,-1]
```

```python
X
```

```python
y
```

```python
class CustomDataset(Dataset):
  def __init__(self, X, y):
    self.X = X
    self.y = y

  def __len__(self):
    return self.X.shape[0]

  def __getitem__(self, index):
    return self.X[index], self.y[index]
```

```python
new_dataset = CustomDataset(X, y)
```

```python
dataloader = DataLoader(new_dataset, batch_size = 32, shuffle = True)
```

# Create LSTM Architecture


```python
class LSTM(nn.Module):
  def __init__(self, vocab_size):
    super().__init__()

    self.embedding = nn.Embedding(vocab_size, embedding_dim = 100)
    self.lstm = nn.LSTM(100, 150, batch_first = True)
    self.fc = nn.Linear(150, vocab_size)

  def forward(self,x):
    embedding = self.embedding(x)
    intermediate_hidden_states, (final_hidden_state, final_cell_state) = self.lstm(embedding)
    output = self.fc(final_hidden_state.squeeze(0))

    return output
```

```python
model = LSTM(len(vocab))
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
model = model.to(device)
```

```python
learning_rate = 0.001
epochs = 100
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr = learning_rate)
```

# Training

```python

# training loop

for epoch in range(epochs):
  total_loss = 0

  for x, y in dataloader:
    x, y = x.to(device), y.to(device)

    optimizer.zero_grad()

    output = model(x)

    loss = criterion(output, y)

    loss.backward()

    optimizer.step()

    total_loss += loss.item()

  print(f"Epoch : {epoch+1}  Loss : {total_loss:.4f}")
```

# Prediction

```python
#  Prediction

def predict(model, text, vocab):

  # tokenize
  tokens = word_tokenize(text.lower())

  # text -> indices
  numeric_text = sentence_to_indices(tokens,vocab)

  # padding
  padded_text = [0]*(61 - len(numeric_text)) + numeric_text
  padded_text = torch.tensor(padded_text, dtype = torch.long).unsqueeze(0)

  # Move padded_text to the same device as the model
  padded_text = padded_text.to(device)

  # send to model
  output = model(padded_text)

  # max prob.
  max_logit, index = torch.max(output, dim = 1)

  # fetch token with that index
  token = list(vocab.keys())[index]

  # return the token with the text in sequence
  return text + " " + token
```
